```python
import os
import librosa
import librosa.feature
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
scale_file = "Audio/scale.wav"
```

```python
# Load audio file
scale, Fs = librosa.load(scale_file)
```

```python
# MEL Filter Banks
filter_banks = librosa.filters.mel(n_fft=2048, sr=22050, n_mels=10)
filter_banks.shape
```

```python
# Showing the filter banks
plt.figure(figsize=(25,10))
librosa.display.specshow(filter_banks, sr =Fs, x_axis='linear')
plt.colorbar(format='%+2.f')
plt.show()
```

```python
# Extracting mel spectrogram
mel_spectrogram = librosa.feature.melspectrogram(y=scale, sr=Fs, n_fft=2048, hop_length=512, n_mels=90)
```

```python
mel_spectrogram.shape
```

```python
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(log_mel_spectrogram, x_axis='time', y_axis='mel', sr=Fs)
plt.colorbar(format='%+2.f')
plt.show()
```
